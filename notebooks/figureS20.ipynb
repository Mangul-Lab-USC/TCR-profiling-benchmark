```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../summary_data/original/Time_memory_benchmarks.csv')

df
```

```python
sns.set_style("white") 
sns.set_context("talk") 

fig, ax = plt.subplots(figsize=(10, 10))
sns.lineplot(data=df, x="RNA-Seq_reads", y="CPU_time", hue="Tool", style="Tool", markers=True)
ax.set_ylabel('CPU Time (s)')
ax.set_xlabel('RNA-Seq reads')
sns.despine()

plt.savefig("../figures/supplementary/figureS20a.pdf",bbox_inches="tight")
plt.savefig("../figures/supplementary/figureS20a.png",bbox_inches="tight")

plt.show()
```

```python
sns.set_style("white") 
sns.set_context("talk") 

fig, ax = plt.subplots(figsize=(10, 10))
sns.lineplot(data=df, x="RNA-Seq_reads", y="Memory", hue="Tool", style="Tool", markers=True)
ax.set_ylabel('Memory (GB)')
ax.set_xlabel('RNA-Seq reads')
sns.despine()

plt.savefig("../figures/supplementary/figureS20b.pdf",bbox_inches="tight")
plt.savefig("../figures/supplementary/figureS20b.png",bbox_inches="tight")

plt.show()
```

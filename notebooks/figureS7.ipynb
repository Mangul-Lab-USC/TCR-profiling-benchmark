```python
import seaborn as sns
import sys
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
import numpy as np

import warnings
warnings.simplefilter(action="ignore", category=FutureWarning)
```

```python
df = pd.read_csv("../summary_data/original/all_tools_TRB_hill_numbers.csv")
df
```

## sample01

```python
sample01 = df.loc[(df['Sample'] == 'sample01')]

sns.set_style("white") 
sns.set_context("talk")  
    
fig, ax = plt.subplots(figsize=(5, 5))

sns.lineplot(sample01['Q'],sample01['TCR-Seq'],label='TCR-Seq',color='#956cb4',marker="o")
sns.lineplot(sample01['Q'],sample01['MIXCR'],label='MIXCR',color='#4878d0',marker="o")
sns.lineplot(sample01['Q'],sample01['IMREP'],label='IMREP',color='#ee854a',marker="o")
sns.lineplot(sample01['Q'],sample01['TRUST4'],label='TRUST4',color='#6acc64',marker="o")
sns.lineplot(sample01['Q'],sample01['CATT'],label='CATT',color='#d65f5f',marker="o")
    
ax.legend(['TCR-Seq','MIXCR','IMREP','TRUST4','CATT'])
  
plt.legend(bbox_to_anchor=(0.3,1), frameon=False, ncol=2, handletextpad=0.2, columnspacing=0.8, handlelength=1, prop={'size': 15})
locs, labels = plt.xticks()
ax.set_ylabel('Diversity estimation')
ax.set_xlabel('Q value')
plt.title('sample01',y=1.05,fontsize=15)

# setting label sizes after creation
ax.xaxis.label.set_size(15)
ax.yaxis.label.set_size(15)
ax.tick_params(axis='both', which='major', labelsize=14)
ax.tick_params(axis='both', which='minor', labelsize=14)
sns.despine()
plt.subplots_adjust(wspace=0.5)

plt.savefig("../figures/supplementary/figureS7/figureS7a.pdf",bbox_inches="tight")
plt.savefig("../figures/supplementary/figureS7/figureS7a.png",bbox_inches="tight")

plt.show()
```

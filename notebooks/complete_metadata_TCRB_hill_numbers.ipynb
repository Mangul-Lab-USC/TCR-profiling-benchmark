```python
import pandas as pd
import numpy as np
from functools import reduce
```

```python
df = pd.read_csv("../summary_data/original/all_tools_TRB.csv")
df
```

## TCR-Seq

```python
TCR = df[['Sample','tissue','tissue_type','CDR3','nReads_TCR','total_reads_TCR','frequency_TCR']]
TCR = TCR[TCR.nReads_TCR != 0]

# q=1
TCR['q1'] = -(TCR['frequency_TCR']*np.log(TCR['frequency_TCR']))
q1_TCR = TCR.groupby(['Sample']).agg({'q1':'sum'}).reset_index().rename(columns={'q1':"shannon_index_tool"})
q1_TCR['TCR-Seq'] = np.exp(q1_TCR['shannon_index_tool'])
q1_TCR['Q'] = '1'
q1_TCR = q1_TCR[['Sample', 'Q', 'TCR-Seq']]

# q=2
TCR['q2'] = (TCR['frequency_TCR']**2)
q2_TCR = TCR.groupby(['Sample']).agg({'q2': 'sum'}).reset_index().rename(columns={'q2': "simpson_index_tool"})
q2_TCR['TCR-Seq'] = 1/q2_TCR['simpson_index_tool']
q2_TCR['Q'] = '2'
q2_TCR = q2_TCR[['Sample', 'Q', 'TCR-Seq']]

# q=3
TCR['q3'] = (TCR['frequency_TCR']**3)
q3_TCR = TCR.groupby(['Sample']).agg({'q3': 'sum'}).reset_index().rename(columns={'q3': "q3_sum"})
q3_TCR['TCR-Seq'] = (q3_TCR['q3_sum']**(-1/2))
q3_TCR['Q'] = '3'
q3_TCR = q3_TCR[['Sample', 'Q', 'TCR-Seq']]

# q=4
TCR['q4'] = (TCR['frequency_TCR']**4)
q4_TCR = TCR.groupby(['Sample']).agg({'q4': 'sum'}).reset_index().rename(columns={'q4': "q4_sum"})
q4_TCR['TCR-Seq'] = (q4_TCR['q4_sum']**(-1/3))
q4_TCR['Q'] = '4'
q4_TCR = q4_TCR[['Sample', 'Q', 'TCR-Seq']]

# q=5
TCR['q5'] = (TCR['frequency_TCR']**5)
q5_TCR = TCR.groupby(['Sample']).agg({'q5': 'sum'}).reset_index().rename(columns={'q5': "q5_sum"})
q5_TCR['TCR-Seq'] = (q5_TCR['q5_sum']**(-1/4))
q5_TCR['Q'] = '5'
q5_TCR = q5_TCR[['Sample', 'Q', 'TCR-Seq']]

# q=6
TCR['q6'] = (TCR['frequency_TCR']**6)
q6_TCR = TCR.groupby(['Sample']).agg({'q6': 'sum'}).reset_index().rename(columns={'q6': "q6_sum"})
q6_TCR['TCR-Seq'] = (q6_TCR['q6_sum']**(-1/5))
q6_TCR['Q'] = '6'
q6_TCR = q6_TCR[['Sample', 'Q', 'TCR-Seq']]
```

```python
TCR_combined = pd.concat([q1_TCR, q2_TCR, q3_TCR, q4_TCR, q5_TCR, q6_TCR])
```

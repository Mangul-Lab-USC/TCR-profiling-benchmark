```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../summary_data/original/all_tools_TRB.csv")
df = df.loc[~(df['frequency_TCR'] == 0)]
df
```

```python
# Select 10 clonotypes with the highest frequency for each sample 
df_top10 = pd.DataFrame()

samples = ['sample01','sample02','sample03','sample04','sample05','sample06','sample07','sample08',
           'sample09','sample10','sample11','sample12','sample13','sample14','TCGA-CZ-4862',
           'TCGA-CZ-5463','TCGA-CZ-5985','SRR5233637','SRR5233639']

for sample in samples: 
    df_sample = df.loc[df['Sample'] == sample]
    df_temp_top = df_sample.nlargest(10,'frequency_TCR')
    
    df_top10 = pd.concat([df_top10, df_temp_top], ignore_index=True, sort=False)
```

```python
TCR_abundance = df_top10.groupby(['Sample','class','tissue_type','repertoire_type']).agg({'frequency_TCR':'sum'}).reset_index().rename(columns={'frequency_TCR':"clonotype_abundance"})
TCR_abundance.loc[:,'clonotype_abundance'] = TCR_abundance['clonotype_abundance']*100
TCR_abundance
```

```python
mean = TCR_abundance.groupby(["class"])["clonotype_abundance"].agg(["mean", "std"]) 
display(mean)
```

```python
mean1 = TCR_abundance.groupby(["tissue_type"])["clonotype_abundance"].agg(["mean", "std"]) 
display(mean1)
```

```python
mean2 = TCR_abundance.groupby(["repertoire_type"])["clonotype_abundance"].agg(["mean", "std"]) 
display(mean2)
```

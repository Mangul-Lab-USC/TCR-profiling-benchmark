```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
samples = ['ESO1-sorted-T-cells_S13_L007','INY1-sorted-T-cells_S14_L007','INY2-sorted-T-cells_S15_L007','MP-11-28-12RNA_S2','SAR-11-14-12RNA_S1','TR2-PBMC_S12','SRR5233639',
           'CMT-baseline1C_CAGATC', 'HM-baseline1C_CGATGT','JSSBaseline-RNA_GTGAAA','LEK-OT110712A_CCGTCC','LEK-baseline_CGATGT','PT0112-B_S3','PT0285-B_S5','PT0310_S9',
            'RAS-baseline_TGACCA','TCGA-CZ-4862','TCGA-CZ-5463','TCGA-CZ-5985','SRR5233637']
```

MIXCR

```python
columns = ['th', 'observed_portion_frequency_aa', 'sample']
df_MIXCR = pd.DataFrame(columns=columns)

for sample in samples: 
    df_temp_aa = pd.read_csv('../summary_data/original/capturing_aa_ability/capturing_aa_ability_MIXCR_'+sample+'.csv')
    df_temp_aa = df_temp_aa[df_temp_aa['observed_portion_frequency'].notna()]
    df_temp_aa = df_temp_aa[df_temp_aa.observed_portion_frequency != 0]
    df_temp_aa.rename(columns={'observed_portion_frequency':'observed_portion_frequency_aa'}, inplace=True)
    
    df_temp_aa['sample'] = sample

    df_MIXCR = pd.concat([df_MIXCR, df_temp_aa], ignore_index=True, sort=False)
    df_MIXCR['tool'] = 'MIXCR'

df_MIXCR
```

IMREP

```python
columns = ['th', 'observed_portion_frequency_aa', 'sample']
df_IMREP = pd.DataFrame(columns=columns)

for sample in samples: 
    df_temp_aa = pd.read_csv('../summary_data/original/capturing_aa_ability/capturing_aa_ability_IMREP_'+sample+'.csv')
    df_temp_aa = df_temp_aa[df_temp_aa['observed_portion_frequency'].notna()]
    df_temp_aa = df_temp_aa[df_temp_aa.observed_portion_frequency != 0]
    df_temp_aa.rename(columns={'observed_portion_frequency':'observed_portion_frequency_aa'}, inplace=True)
    
    df_temp_aa['sample'] = sample

    df_IMREP = pd.concat([df_IMREP, df_temp_aa], ignore_index=True,sort=False)
    df_IMREP['tool'] = 'IMREP'

df_IMREP
```

TRUST4

```python
columns = ['th', 'observed_portion_frequency_aa', 'sample']
df_TRUST4 = pd.DataFrame(columns=columns)

for sample in samples: 
    df_temp_aa = pd.read_csv('../summary_data/original/capturing_aa_ability/capturing_aa_ability_TRUST4_'+sample+'.csv')
    df_temp_aa = df_temp_aa[df_temp_aa['observed_portion_frequency'].notna()]
    df_temp_aa = df_temp_aa[df_temp_aa.observed_portion_frequency != 0]
    df_temp_aa.rename(columns={'observed_portion_frequency':'observed_portion_frequency_aa'}, inplace=True)
   
    df_temp_aa['sample'] = sample

    df_TRUST4 = pd.concat([df_TRUST4, df_temp_aa], ignore_index=True, sort=False)
    df_TRUST4['tool'] = 'TRUST4'

df_TRUST4
```

```python
df_combined = pd.concat([df_MIXCR,df_IMREP,df_TRUST4], ignore_index=True, sort=False)
df_combined['th'] = df_combined['th']
df_combined
```

```python
T_cell_rich_monoclonal = df_combined[df_combined['sample'].isin(['ESO1-sorted-T-cells_S13_L007','INY1-sorted-T-cells_S14_L007','INY2-sorted-T-cells_S15_L007'])]
    
tools = ['MIXCR','IMREP','TRUST4']
colors = ['#4878d0', '#ee854a', '#6acc64']
samples = ['ESO1-sorted-T-cells_S13_L007','INY1-sorted-T-cells_S14_L007','INY2-sorted-T-cells_S15_L007']

for sample in samples: 
    sample_figure = T_cell_rich_monoclonal.loc[T_cell_rich_monoclonal['sample'] == sample]
    
    for tool,color in zip (tools,colors):

        sns.set_style("white") 
        sns.set_context("talk")  
    
        T_cell_rich_monoclonal_figure = sample_figure.loc[sample_figure['tool'] == tool]
    
        fig, ax = plt.subplots(figsize=(5, 4))
        ax.stackplot(T_cell_rich_monoclonal_figure['th'], T_cell_rich_monoclonal_figure['observed_portion_frequency_aa'],labels=['aa'],colors=color)
    
        locs, labels = plt.xticks()
        ax.set_ylabel('Portion of repetoire captured by' + tool)
        ax.set_xlabel('Clonotype frequency')
        plt.title(sample,y=1.2,fontsize=16)
        ax.set_xlim(0,0.0004)

        # setting label sizes after creation
        ax.xaxis.label.set_size(16)
        ax.yaxis.label.set_size(16)
        ax.tick_params(axis='both', which='major', labelsize=13)
        ax.tick_params(axis='both', which='minor', labelsize=13)
        plt.xticks(rotation=45)
        sns.despine()
        plt.subplots_adjust(wspace=0.5)
        plt.show(block=False)
```

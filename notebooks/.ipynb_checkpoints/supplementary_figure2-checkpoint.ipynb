```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.ticker as mtick
```

```python
df = pd.read_csv("../summary_data/original/all_tools_TRB.csv")
df = df.loc[~(df['frequency_TCR'] == 0)]
df
```

```python
# Define clonotype groups based on frequency
def clonotype_group (row):
    if row['frequency_TCR'] > 0.01 and row['frequency_TCR'] <= 1:
        return 'Hyperexpanded'
    if row['frequency_TCR'] > 0.001 and row['frequency_TCR'] <= 0.01: 
        return 'Large'
    if row['frequency_TCR'] > 0.0001 and row['frequency_TCR'] <= 0.001:
        return 'Medium'
    if row['frequency_TCR'] > 0.00001 and row['frequency_TCR'] <= 0.0001:
        return 'Small'
    if row['frequency_TCR'] > 0 and row['frequency_TCR'] <= 0.00001:
        return 'Rare'

# Apply the clonotype_group function to the dataframe
df['clonotype_group'] = df.apply (lambda row: clonotype_group(row),axis=1)
df
```

Cumulative clonotype frequency for each clonotype group in each sample

```python
TCR_abundance= df.groupby(['Sample','tissue_type','repertoire_type','clonotype_group']).agg({'frequency_TCR':'sum'}).reset_index().rename(columns={'frequency_TCR':"clonotype_abundance"})
TCR_abundance.loc[:,'clonotype_abundance']=TCR_abundance['clonotype_abundance']*100
TCR_abundance.head(20)
```

```python
# Bar plot per sample
samples = ['ESO1-sorted-T-cells_S13_L007','INY1-sorted-T-cells_S14_L007','INY2-sorted-T-cells_S15_L007','MP-11-28-12RNA_S2','SAR-11-14-12RNA_S1','TR2-PBMC_S12',
           'CMT-baseline1C_CAGATC', 'HM-baseline1C_CGATGT','JSSBaseline-RNA_GTGAAA','LEK-OT110712A_CCGTCC','LEK-baseline_CGATGT','PT0112-B_S3','PT0285-B_S5','PT0310_S9',
            'RAS-baseline_TGACCA','TCGA-CZ-4862','TCGA-CZ-5463','TCGA-CZ-5985','SRR5233639','SRR5233637']
label_order = ['Hyperexpanded','Large','Medium','Small','Rare']
palette = sns.color_palette("muted")
palette.reverse()

for sample in samples:
    sample_clonotype_abundance = TCR_abundance.loc[(TCR_abundance['Sample'] == sample)]
    plt.figure(figsize=(5,5))
    ax = sns.barplot(x='clonotype_group', y='clonotype_abundance', data=sample_clonotype_abundance, order=label_order, palette=palette)
    ax.set_xlabel('Clonotype group',fontsize=10)
    ax.set_ylabel('Clonotype frequency',fontsize=10)
    ax.set(title = sample)
    ax.yaxis.set_major_formatter(mtick.PercentFormatter())
    ax.set_ylim(0,100)
    plt.xticks(fontsize=10,rotation=45)
    plt.yticks(fontsize=10)

    plt.show()
```

Clonotype count for each clonotype group in samples

```python
#TCR
TCR = df.groupby(['Sample','tissue_type','repertoire_type','clonotype_group']).size().reset_index(name='clonotype_group_count')
TCR['tool'] = 'TCR-Seq'

# MIXCR
MIXCR = df[df['frequency_MIXCR'] != 0].groupby(['Sample','tissue_type','repertoire_type','clonotype_group'])['frequency_MIXCR'].count().reset_index(name='clonotype_group_count')
MIXCR['tool'] = 'MIXCR'

# IMREP
IMREP = df[df['frequency_IMREP'] != 0].groupby(['Sample','tissue_type','repertoire_type','clonotype_group'])['frequency_IMREP'].count().reset_index(name='clonotype_group_count')
IMREP['tool'] = 'IMREP'

# TRUST4
TRUST4 = df[df['frequency_TRUST4'] != 0].groupby(['Sample','tissue_type','repertoire_type','clonotype_group'])['frequency_TRUST4'].count().reset_index(name='clonotype_group_count')
TRUST4['tool'] = 'TRUST4'

combined_count = pd.concat([TCR, MIXCR, IMREP, TRUST4],ignore_index=True)
combined_count.head(20)
```

```python
# Bar plot per sample
samples = ['ESO1-sorted-T-cells_S13_L007','INY1-sorted-T-cells_S14_L007','INY2-sorted-T-cells_S15_L007','MP-11-28-12RNA_S2','SAR-11-14-12RNA_S1','TR2-PBMC_S12',
           'CMT-baseline1C_CAGATC', 'HM-baseline1C_CGATGT','JSSBaseline-RNA_GTGAAA','LEK-OT110712A_CCGTCC','LEK-baseline_CGATGT','PT0112-B_S3','PT0285-B_S5','PT0310_S9',
            'RAS-baseline_TGACCA','TCGA-CZ-4862','TCGA-CZ-5463','TCGA-CZ-5985','SRR5233639','SRR5233637']
label_order = ['Hyperexpanded','Large','Medium','Small','Rare']

for sample in samples:
    sample_clonotype_count = combined_count.loc[(combined_count['Sample'] == sample)]
    plt.figure(figsize=(5,5))
    ax = sns.barplot(x='clonotype_group', y='clonotype_group_count', hue='tool', hue_order= ['MIXCR','IMREP','TRUST4','TCR-Seq'], data=sample_clonotype_count, order=label_order, palette='muted')
    ax.set_yscale("log")
    ax.set_xlabel('Clonotype group',fontsize=10)
    ax.set_ylabel('Clonotype count (log)',fontsize=10)
    ax.set(title = sample)
    plt.xticks(fontsize=10,rotation=45)
    plt.yticks(fontsize=10)

    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
import numpy as np
```

```python
# load dataframe
df_V = pd.read_csv("../summary_data/150bp/all_tools_TRB_V_usage.csv")
df_J = pd.read_csv("../summary_data/150bp/all_tools_TRB_J_usage.csv")
```

```python
V_count = df_V.groupby(['tool','sample'])['V'].nunique().reset_index()

order_list = ['IMREP','MIXCR','TRUST4','TCR-Seq']
ax = sns.barplot(x='tool',y='V',data=V_count,order=order_list)
ax.set_xlabel('tool',fontsize=10)
ax.set_ylabel('Number of V gene',fontsize=10)
plt.xticks(rotation=90)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.show()
```

```python
plt.figure(figsize=(15,16))
ax = sns.barplot(x='V', y='frequency_tool', hue='tool', data=df_V, errwidth=1)
ax.set_xlabel('V gene',fontsize=10)
ax.set_ylabel('Gene frequency',fontsize=10)
plt.xticks(rotation=90)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.show()
```

```python
tools = ['IMREP','MIXCR','TRUST4','TCR-Seq']

for tool in tools:
    figure_V = df_V.loc[(df_V['tool'] == tool)]
    ax = sns.barplot(x='V', y='frequency_tool', data=figure_V)
    ax.set(title = tool)
    ax.set_xlabel('V gene',fontsize=10)
    ax.set_ylabel('Gene frequency',fontsize=10)
    plt.xticks(rotation=90)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)

    plt.show()
```

```python
# reshape the dataframe for correlation analysis
TCR_V = df_V.loc[(df_V['tool'] == 'TCR-Seq')]
TCR_V.columns = ['sample', 'V', 'frequency_TCR', 'tool', 'sample_type', 'clonal_type']
TCR_V = TCR_V.drop('tool', 1)
TCR_V

tool_V = df_V[(df_V['tool'] != 'TCR-Seq')]
df_V_correlation = pd.merge(tool_V, TCR_V, how='inner', on=['sample','V','sample_type','clonal_type'])
df_V_correlation
```

```python
tools = ['MIXCR','IMREP','TRUST4']

for tool in tools: 
    figure_V_correlation = df_V_correlation.loc[(df_V_correlation['tool'] == tool)]
    sns.scatterplot(x = 'frequency_TCR', y = 'frequency_tool', hue = 'clonal_type', s = 25, data = figure_V_correlation)
    ax = sns.lineplot(x=range(0,2),y=range(0,2))
    ax.set(title = tool)
    ax.lines[0].set_linestyle("--")
    ax.set(xlim = (0, 1), ylim = (0, 1))
    ax.grid(False)
    plt.show()
```

```python
# calculate the Pearson correlation 
tools = ['MIXCR','IMREP','TRUST4']              
for tool in tools: 
    correlation = df_V_correlation.loc[(df_V_correlation['tool'] == tool)]
    print(tool + ' Pearson correlation cofficient:' + str(stats.pearsonr(correlation['frequency_TCR'], correlation['frequency_tool'])))
```

```python
# calculate the Pearson correlation for polyclonal samples
polyclonal_V = df_V_correlation.loc[(df_V_correlation['clonal_type'] == 'polyclonal')]

tools = ['MIXCR','IMREP','TRUST4']              
for tool in tools: 
    correlation = polyclonal_V.loc[(polyclonal_V['tool'] == tool)]
    print(tool + ' Pearson correlation cofficient:' + str(stats.pearsonr(correlation['frequency_TCR'], correlation['frequency_tool'])))
```

```python
# calculate the Pearson correlation for monoclonal samples
monoclonal_V = df_V_correlation.loc[(df_V_correlation['clonal_type'] == 'monoclonal')]

tools = ['MIXCR','IMREP','TRUST4']              
for tool in tools: 
    correlation = monoclonal_V.loc[(monoclonal_V['tool'] == tool)]
    print(tool + ' Pearson correlation cofficient:' + str(stats.pearsonr(correlation['frequency_TCR'], correlation['frequency_tool'])))
```
